```python
import json
import math
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
from transformers import AutoTokenizer
from rich.console import Console
from rich.table import Table
from rich.text import Text
from rich.theme import Theme
from rich.panel import Panel

# ==========================================
# 0. Rich UI 設定與參數配置
# ==========================================
MAX_SEQ_LENGTH = 512  # 你預計訓練時設定的最大序列長度

custom_theme = Theme({
    "info": "dim cyan",
    "warning": "magenta",
    "danger": "bold red",
    "success": "bold green",
    "im_start": "bold bright_cyan",
    "im_end": "bold bright_red",
    "think_tag": "bold yellow",
    "end_think_tag": "bold gold3",
    "final_tag": "bold green",
    "end_final_tag": "bold dark_green"
})
console = Console(theme=custom_theme)

# ==========================================
# 1. 設定與初始化
# ==========================================
json_file = "/Users/hungwei/Desktop/Proj/Mamba3-XR/data/stf.json" 
model_name = "/Users/hungwei/Desktop/Proj/Mamba3-XR/data/"

special_tokens = ["<|im_start|>", "<|im_end|>", "<think>", "</think>", "<final>", "</final>"]

# 常見推理邏輯連接詞 (可用於評估思考深度)
logical_connectors = ["因此", "因為", "假設", "步驟", "therefore", "assuming", "because", "so", "step"]

console.print(Panel(f"正在初始化 Tokenizer...\n模型路徑: [bold]{model_name}[/bold]", title="Step 1", style="info"))
tokenizer = AutoTokenizer.from_pretrained(model_name)
tokenizer.add_tokens(special_tokens)

# ==========================================
# 2. 讀取與精細 Tokenize 統計 (加入新模組)
# ==========================================
console.print(Panel(f"正在讀取並分析數據內容...\n資料路徑: [bold]{json_file}[/bold]", title="Step 2", style="info"))
with open(json_file, "r", encoding="utf-8") as f:
    data = json.load(f)

analysis_results = []
formatted_texts = []
format_errors = [] # 儲存標籤配對錯誤的資料索引與原因

# 用來儲存第一筆資料的底層細節
first_sample_ids = []
first_sample_tokens = []

for i, item in enumerate(data):
    user_input = item.get("input", "").strip()
    cot_content = item.get("cot", "").strip()
    final_output = item.get("output", "").strip()
    category = item.get("category", "Uncategorized")

    # [計算 1] 分段計算 Token
    input_ids = tokenizer.encode(user_input, add_special_tokens=False)
    cot_ids = tokenizer.encode(cot_content, add_special_tokens=False)
    output_ids = tokenizer.encode(final_output, add_special_tokens=False)

    # 完整格式化文本
    formatted = (
        f"<|im_start|>user\n{user_input}\n<|im_end|>\n"
        f"<|im_start|>assistant\n"
        f"<think>\n{cot_content}\n</think>\n"
        f"<final>\n{final_output}\n</final>\n"
        f"<|im_end|>\n"
    )
    formatted_texts.append(formatted)
    
    # [計算 2] 總長度與訓練有效比例 (Loss Masking Ratio)
    full_ids = tokenizer.encode(formatted) 
    trainable_tokens = len(cot_ids) + len(output_ids)
    trainable_ratio = trainable_tokens / max(len(full_ids), 1)

    # [計算 3] CoT 思考步驟預估 (Reasoning Steps)
    # 計算換行數、數字條目 (如 1. 2.) 以及邏輯詞出現次數
    newline_count = cot_content.count('\n')
    list_item_count = len(re.findall(r'(?:^|\n)\s*\d+\.\s', cot_content))
    connector_count = sum(cot_content.count(c) for c in logical_connectors)
    estimated_steps = newline_count + list_item_count + (connector_count // 2)

    # [計算 4] 特殊標籤完整性檢查 (Sanity Checking)
    error_flags = []
    if formatted.count("<think>") != formatted.count("</think>"):
        error_flags.append("Think_Tag_Mismatch")
    if formatted.count("<final>") != formatted.count("</final>"):
        error_flags.append("Final_Tag_Mismatch")
    if formatted.count("<|im_start|>") != formatted.count("<|im_end|>"):
        error_flags.append("ChatML_Tag_Mismatch")
    
    if error_flags:
        format_errors.append({"index": i, "errors": error_flags})

    # 擷取第一筆資料的詳細 Token 資訊
    if i == 0:
        first_sample_ids = full_ids
        first_sample_tokens = tokenizer.convert_ids_to_tokens(full_ids)

    analysis_results.append({
        "category": category,
        "input_tokens": len(input_ids),
        "cot_tokens": len(cot_ids),
        "output_tokens": len(output_ids),
        "total_tokens": len(full_ids),
        "cot_ratio": len(cot_ids) / (len(cot_ids) + len(output_ids) + 1e-9),
        "trainable_ratio": trainable_ratio,
        "reasoning_steps": estimated_steps
    })

df = pd.DataFrame(analysis_results)
total_dataset_tokens = df['total_tokens'].sum() 

# ==========================================
# 3. 訓練策略、Packing 與品質報告
# ==========================================
console.print("\n[bold cyan]正在生成訓練評估與資料品質報告...[/bold cyan]\n")

oversized_count = (df['total_tokens'] > MAX_SEQ_LENGTH).sum()
oversized_ratio = (oversized_count / len(df)) * 100
packed_sequences = math.ceil(total_dataset_tokens / MAX_SEQ_LENGTH)
low_trainable_count = (df['trainable_ratio'] < 0.3).sum() # 學習比例低於 30% 的資料數

advice_text = (
    f"[bold]1. 資料集總 Token 量[/bold]: {total_dataset_tokens:,} tokens\n"
    f"[bold]2. 設定的最大長度 (seq_len)[/bold]: {MAX_SEQ_LENGTH}\n\n"
    f"[bold]3. 截斷風險 (Truncation)[/bold]:\n"
    f"   有 [danger]{oversized_count}[/danger] 筆資料 ({oversized_ratio:.1f}%) 超過 {MAX_SEQ_LENGTH} tokens。\n\n"
    f"[bold]4. 訓練有效比例 (Loss Masking)[/bold]:\n"
    f"   全集平均有效學習比例為 [warning]{df['trainable_ratio'].mean()*100:.1f}%[/warning]。(目標：Assistant 回應占比越高越好)\n"
    f"   有 [danger]{low_trainable_count}[/danger] 筆資料的有效學習 Token 佔比不到 30%，可能拖慢訓練效率。\n\n"
    f"[bold]5. 拼接策略 (Token Packing)[/bold]:\n"
    f"   開啟 Packing 後，大約會將 {len(df)} 筆資料壓縮成 [success]{packed_sequences}[/success] 個訓練 Steps。"
)
console.print(Panel(advice_text, title="🧠 SFT 訓練與 Packing 建議", border_style="cyan"))

# 印出標籤檢查結果
if format_errors:
    error_msg = f"[bold red]發現 {len(format_errors)} 筆資料有標籤閉合錯誤！(無限生成風險)[/bold red]\n"
    for err in format_errors[:5]: # 最多印出前5筆避免洗頻
        error_msg += f"- 第 {err['index']} 筆: {', '.join(err['errors'])}\n"
    if len(format_errors) > 5:
        error_msg += f"...等共 {len(format_errors)} 筆"
    console.print(Panel(error_msg, title="🚨 資料格式警告 (Sanity Check)", border_style="red"))
else:
    console.print(Panel("[bold green]完美！所有資料的特殊標籤 (<think>, <final>, ChatML) 皆成功閉合配對。[/bold green]", title="✅ 資料格式檢查", border_style="green"))

# ==========================================
# 4. 第一筆資料深度解剖
# ==========================================
console.print("\n[bold u]🔬 第一筆資料深度解剖 (檢查特殊 Token 是否正確)[/bold u]\n")

escaped_tokens = [re.escape(t) for t in special_tokens]
pattern = re.compile(f"({'|'.join(escaped_tokens)})")
parts = pattern.split(formatted_texts[1028])
styled_text = Text()
for part in parts:
    style = custom_theme.styles.get(custom_theme.styles.get(part, "white"), "white")
    if part in special_tokens:
        styled_text.append(part, style="bold bright_cyan")
    else:
        styled_text.append(part, style="white")

console.print(Panel(styled_text, title="1. 格式化後的純文字 (Text)", border_style="dim"))

ids_str = ", ".join(map(str, first_sample_ids))
if len(first_sample_ids) > 100:
    ids_str = ", ".join(map(str, first_sample_ids[:50])) + " ... [省略] ... " + ", ".join(map(str, first_sample_ids[-10:]))
console.print(Panel(ids_str, title=f"2. 模型看到的數字 (Input IDs) - 共 {len(first_sample_ids)} 個", border_style="yellow"))

tokens_str = " | ".join(first_sample_tokens)
if len(first_sample_tokens) > 100:
    tokens_str = " | ".join(first_sample_tokens[:50]) + " ... [省略] ... " + " | ".join(first_sample_tokens[-10:])
console.print(Panel(tokens_str, title="3. ID 對應的切詞結果 (Tokens)", border_style="magenta"))

# ==========================================
# 5. 視覺化分析圖表 (3x3 進階擴展版面)
# ==========================================
console.print("\n[bold cyan]正在生成進階視覺化圖表視窗... (請查看彈出的圖片)[/bold cyan]")

plt.style.use('seaborn-v0_8-muted')
fig, axes = plt.subplots(3, 3, figsize=(24, 18)) 

# [圖 1] 類別比例圓餅圖
df['category'].value_counts().plot(kind='pie', autopct='%1.1f%%', ax=axes[0, 0], startangle=140, cmap='Pastel1')
axes[0, 0].set_title("1. Category Distribution")
axes[0, 0].set_ylabel("")

# [圖 2] 總 Token 分佈直方圖
sns.histplot(df['total_tokens'], bins=30, kde=True, color='purple', ax=axes[0, 1])
axes[0, 1].axvline(df['total_tokens'].mean(), color='red', linestyle='--', label=f"Mean: {df['total_tokens'].mean():.1f}")
axes[0, 1].axvline(MAX_SEQ_LENGTH, color='black', linestyle='-.', linewidth=2, label=f"Max Seq ({MAX_SEQ_LENGTH})")
axes[0, 1].set_title("2. Total Token Distribution")
axes[0, 1].legend()

# [圖 3] 分段 Token 分佈疊加圖
sns.kdeplot(df['input_tokens'], fill=True, label="Input", ax=axes[0, 2])
sns.kdeplot(df['cot_tokens'], fill=True, label="CoT", ax=axes[0, 2])
sns.kdeplot(df['output_tokens'], fill=True, label="Output", ax=axes[0, 2])
axes[0, 2].set_title("3. Segment Density")
axes[0, 2].legend()

# [圖 4] 類別與長度的箱型圖
sns.boxplot(x='category', y='total_tokens', data=df, ax=axes[1, 0])
axes[1, 0].axhline(MAX_SEQ_LENGTH, color='red', linestyle='-.', alpha=0.5)
axes[1, 0].set_title("4. Total Tokens by Category")
axes[1, 0].tick_params(axis='x', rotation=45)

# [圖 5] Input vs CoT 相關性
sns.scatterplot(x='input_tokens', y='cot_tokens', hue='category', data=df, ax=axes[1, 1], s=80, alpha=0.7)
axes[1, 1].set_title("5. Input vs CoT Correlation")

# [圖 6] CoT 佔比分佈圖
sns.histplot(df['cot_ratio'] * 100, bins=20, kde=True, color='teal', ax=axes[1, 2])
axes[1, 2].axvline(df['cot_ratio'].mean() * 100, color='red', linestyle='--', label=f"Mean: {df['cot_ratio'].mean()*100:.1f}%")
axes[1, 2].set_title("6. CoT Ratio Distribution (%)")
axes[1, 2].legend()

# [圖 7] (新增) 訓練有效比例分佈 (Trainable Ratio)
sns.histplot(df['trainable_ratio'] * 100, bins=20, kde=True, color='coral', ax=axes[2, 0])
axes[2, 0].axvline(df['trainable_ratio'].mean() * 100, color='red', linestyle='--', label=f"Mean: {df['trainable_ratio'].mean()*100:.1f}%")
axes[2, 0].axvline(30, color='black', linestyle='-.', label="30% Threshold") # 標示 30% 警告線
axes[2, 0].set_title("7. Trainable Ratio (Loss Masking) (%)")
axes[2, 0].legend()

# [圖 8] (新增) 預估推理步驟數分佈
sns.boxplot(x='category', y='reasoning_steps', data=df, ax=axes[2, 1], palette="Set2")
axes[2, 1].set_title("8. Estimated Reasoning Steps by Category")
axes[2, 1].tick_params(axis='x', rotation=45)

# [圖 9] (新增) 預估推理步驟 vs CoT Tokens
sns.scatterplot(x='reasoning_steps', y='cot_tokens', hue='category', data=df, ax=axes[2, 2], s=80, alpha=0.7)
axes[2, 2].set_title("9. Reasoning Steps vs CoT Token Length")

plt.tight_layout()
plt.show()
```
